```python
from skimage.io import imread
import matplotlib.pyplot as plt
from functionality import *
```

```python
K = np.arange(3)
n_labels = len(K)
```

```python
# img1
img = imread('test_images/img1.jpg').astype(int)[...,:3]
height, width, _ = img.shape
plt.axis('off')
plt.imshow(img)
```

```python
# img1
beta = 2 
bbox0 =   0, 38, 0, width
bbox1 =  50, 100, 0, 160
bbox2 =  100, height, 0, width
```

```python
mean0, cov0, mean1, cov1, mean2, cov2 = get_gmm_params(img, bbox2, bbox1, bbox0)
g = get_g(height,width,n_labels,beta)
Q = get_q(img,mean0, cov0, mean1, cov1, mean2, cov2)
```

```python
fi = np.zeros((height,width,4,n_labels))
for _ in range(10**3):
    print(_)
    fi = update_fi(height,width,fi,g,K,Q,10)
    labelling = get_labelling(height, width, g, fi)
    smooth_border, Q  = check_smooth_border(labelling,Q)
    if not smooth_border:
        smooth_border, Q = check_cons(labelling,Q)
    plt.imshow(labelling)
    plt.axis('off')
    plt.show()
    if not smooth_border:
        break
```
